# Importing the required libraries

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Set visualization style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

# Enable inline plotting for Jupyter notebooks
%matplotlib inline
```

# Data Loading and Initial Exploration

```python
# Load the dataset
df = pd.read_csv('Superstore.csv', encoding='latin1')
```

```python
# Display basic information
print("Dataset Shape:", df.shape)
print("\nFirst 5 rows:")
display(df.head())

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nSummary Statistics:")
display(df.describe(include='all'))
```

# Data Cleaning and Preprocessing

```python
# Convert date columns to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m/%d/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%m/%d/%Y')

# Extract year, month, and day from Order Date
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Day'] = df['Order Date'].dt.day
df['Order Weekday'] = df['Order Date'].dt.day_name()

# Calculate time to ship
df['Days to Ship'] = (df['Ship Date'] - df['Order Date']).dt.days

# Check for negative days to ship (data quality issue)
if (df['Days to Ship'] < 0).any():
    print("Warning: Negative days to ship found!")
    
# Check for duplicates
print("\nNumber of duplicate rows:", df.duplicated().sum())
```

## RFM ( Recency , Frequency, Monetory) Analysis

```python
# Calculate RFM metrics
snapshot_date = df['Order Date'].max() + pd.Timedelta(days=1)

rfm = df.groupby('Customer ID').agg({
    'Order Date': lambda x: (snapshot_date - x.max()).days,
    'Order ID': 'count',
    'Sales': 'sum'
})

rfm.rename(columns={
    'Order Date': 'Recency',
    'Order ID': 'Frequency',
    'Sales': 'MonetaryValue'
}, inplace=True)

# Create RFM segments
r_labels = range(4, 0, -1)
f_labels = range(1, 5)
m_labels = range(1, 5)

r_quartiles = pd.qcut(rfm['Recency'], q=4, labels=r_labels)
f_quartiles = pd.qcut(rfm['Frequency'], q=4, labels=f_labels)
m_quartiles = pd.qcut(rfm['MonetaryValue'], q=4, labels=m_labels)

rfm = rfm.assign(R=r_quartiles, F=f_quartiles, M=m_quartiles)

# Calculate RFM Score
rfm['RFM_Score'] = rfm[['R', 'F', 'M']].sum(axis=1)

# Create segments
def get_rfm_segment(df):
    if df['RFM_Score'] >= 9:
        return 'Champions'
    elif df['RFM_Score'] >= 6:
        return 'Potential Loyalists'
    elif df['RFM_Score'] >= 5:
        return 'At Risk Customers'
    else:
        return 'Lost Customers'

rfm['Segment'] = rfm.apply(get_rfm_segment, axis=1)

# Visualize segments
plt.figure(figsize=(10, 6))
rfm['Segment'].value_counts().plot(kind='bar', color='darkblue')
plt.title('Customer Segments')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.savefig('RFM-Analysis.png',dpi = 300 ,bbox_inches = 'tight')
plt.show()
```

### 🔍 Key Insights from Customer Segmentation (RFM Analysis)

- **Champions (≈300 customers):** These are your most valuable and engaged customers. They purchase frequently, spend the most, and are the best candidates for loyalty and referral programs.
- **Potential Loyalists (≈290 customers):** This segment shows high potential to become Champions. Consider nurturing


## Shipping mode analysis

```python

shipping_analysis = df.groupby('Ship Mode').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Days to Ship': 'mean'
}).sort_values('Sales', ascending=False)

print("\nShipping Mode Analysis:")
display(shipping_analysis)

# Shipping time vs. profit
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Days to Ship', y='Profit', hue='Ship Mode')
plt.title('Shipping Time vs. Profit')
plt.savefig('Shipping Mode Analysis.png',dpi = 300 ,bbox_inches = 'tight')
plt.show()
```

###  Key Insights from Shipping Mode Analysis

- **Standard Class** is the most widely used shipping mode, accounting for the highest **sales ($1.36M)** and **profit (~$164K)**. However, it also has the **longest average shipping time (5 days)**.
- **Second Class** offers moderate performance, generating **$459K in sales** and **~$57K in profit**, with an average **shipping time of ~3.2 days**.
- **First Class** shows relatively high profitability (**~$49K**) despite lower sales volume (**$351K**), and provides faster delivery (**~2.2 days**).
- **Same Day shipping**, while the fastest (**~0 days**), contributes the least in terms of **sales ($128K)** and **profit (~$15.9K)**, possibly due to limited availability or higher operational costs.

> **Insight:** There's a clear trade-off between shipping speed and overall profitability. While Standard Class has the slowest delivery, it remains the most profitable. Businesses could explore expanding First Class services to improve customer satisfaction without severely impacting margins.
